```python
import librosa  # Audio loading, effects, and resampling wizard
import numpy as np  # Math helper for arrays and peaks
import soundfile as sf  # For saving the processed .wav (high-quality)

def preprocess_audio(input_path, output_path=None, target_sr=16000, trim_db=20, normalize=True):

    print(f"üîÑ Preprocessing: {input_path}")
    
    # Step 1: Load the Original Audio (Raw Waveform + Sample Rate)
    # Why? Captures everything‚Äîstereo, original rate (e.g., 22kHz for GTZAN).
    try:
        y_original, sr_original = librosa.load(input_path, sr=None, mono=False)  # mono=False keeps stereo
        print(f"   üìä Loaded: Duration={len(y_original)/sr_original:.2f}s, Channels={'Stereo' if y_original.ndim > 1 else 'Mono'}, SR={sr_original}Hz")
    except Exception as e:
        raise ValueError(f"‚ùå Load failed for {input_path}: {e}")
    
    # Step 2: Resample to 16kHz & Convert to Mono
    # Why? 16kHz is efficient (cuts samples ~27% from 22kHz); mono averages L/R channels for simplicity.
    # Librosa handles resampling smoothly (anti-aliasing filters prevent distortion).
    y_mono = librosa.resample(y_original, orig_sr=sr_original, target_sr=target_sr) if y_original.ndim > 1 else librosa.resample(y_original, orig_sr=sr_original, target_sr=target_sr)
    if y_original.ndim > 1:  # Stereo? Average channels to mono
        y_mono = np.mean(y_mono, axis=0)  # Shape: (samples,) now
    sr_processed = target_sr
    print(f"   üîÑ Resampled to {sr_processed}Hz Mono: New duration={len(y_mono)/sr_processed:.2f}s")
    
    # Step 3: Trim Silence (Remove Long Quiet Starts/Ends)
    # Why? GTZAN files often have fade-ins/outs‚Äîtrimming focuses on content, shortens clips.
    # top_db=20 means "trim where amplitude < -20dB" (quiet threshold‚Äîtune lower for aggressive).
    y_trimmed, trim_info = librosa.effects.trim(y_mono, top_db=trim_db)
    print(f"   ‚úÇÔ∏è Trimmed silence: From {len(y_mono)/sr_processed:.2f}s to {len(y_trimmed)/sr_processed:.2f}s (indices {trim_info})")
    
    # Step 4: Normalize Amplitude (Scale to Peak [-1, 1])
    # Why? Ensures all tracks have similar volume‚Äîkey for fair feature extraction (e.g., MFCCs won't bias loud songs).
    # np.max(np.abs()) finds peak abs value, divides to cap at 1 (preserves shape, just scales).
    if normalize:
        peak = np.max(np.abs(y_trimmed))
        if peak > 0:  # Avoid divide-by-zero (silent files)
            y_normalized = y_trimmed / peak
        else:
            y_normalized = y_trimmed  # Already zero‚Äîrare
        print(f"   üìê Normalized: Peak from {peak:.3f} to 1.0")
    else:
        y_normalized = y_trimmed
    
    # Step 5: Reduce Silence (Gentle Compression for Background Noise)
    # Why? Trimming cuts edges, but mid-track hums/noise? This quiets them without full removal.
    # Simple compressor: Boost quiet parts less, squash loud peaks (ratio=4:1, threshold=-20dB).
    # (Advanced: Librosa's compressor is basic; this is a lightweight NumPy version.)
    threshold_db = -20  # Quiet below this dB gets reduced
    ratio = 4  # Compression ratio (e.g., 4dB over threshold ‚Üí 1dB output)
    
    # Convert to dB for compression logic
    y_db = librosa.amplitude_to_db(np.abs(y_normalized), ref=np.max)
    
    # Apply compression: For parts > threshold, reduce gain
    over_threshold = y_db > threshold_db
    compression_gain = 1 / ratio * over_threshold + (1 - over_threshold)  # Scale down loud parts
    y_compressed = y_normalized * 10**(compression_gain / 20)  # Back to linear amplitude
    
    print(f"   üîâ Reduced silence: Applied {ratio}:1 compression above {threshold_db}dB")
    
    y_processed = y_compressed  # Final output
    
    # Step 6: Save if Requested (High-Quality .wav)
    if output_path:
        sf.write(output_path, y_processed, sr_processed)
        print(f"   üíæ Saved: {output_path}")
    
    print(f"‚úÖ Done! Processed shape: {y_processed.shape}, SR: {sr_processed}Hz")
    return y_processed, sr_processed

# Demo: Test on Your Sample File (From Earlier Conversation)
if __name__ == "__main__":
    # Your paths from before‚Äîadjust as needed
    sample_input = r'C:\Users\Akil S\OneDrive\Desktop\infosys\archive\Data\genres_original\blues\blues.00000.wav'
    sample_output = 'blues_processed_16kHz.wav'  # Saves in current folder
    
    # Run the magic!
    y_proc, sr_proc = preprocess_audio(sample_input, sample_output, target_sr=16000)
    
    # Bonus: Quick Stats & Plot (Visual Check)
    print(f"\nüìà Final Stats:")
    print(f" - Amplitude range: [{np.min(y_proc):.3f}, {np.max(y_proc):.3f}] (normalized!)")
    print(f" - Zero-crossings (activity): ~{np.sum(np.diff(np.sign(y_proc)) != 0)/len(y_proc)*100:.1f}% non-silent")
    
    import matplotlib.pyplot as plt
    plt.figure(figsize=(10, 4))
    time_axis = np.linspace(0, len(y_proc)/sr_proc, len(y_proc))
    plt.plot(time_axis, y_proc)
    plt.title('Processed Audio: 16kHz Mono, Normalized, Silence Handled')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()
```
