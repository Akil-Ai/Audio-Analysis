# 1. CLASS AND OBJECT
## Example

```python

class Car:
    def __init__(self, brand):
        self.brand = brand

# Class: Car, Object: my_car
my_car = Car("Toyota")
print(f"My {my_car.brand}")  
```

# 2. ENCAPSULATION
## Example 1

```python
class AudioFile:
    def __init__(self, name, duration):
        self.__name = name          # private
        self.__duration = duration  # private

    def get_name(self):
        return self.__name

    def get_duration(self):
        return self.__duration
```

# 3. Inheritance

```python

class Animal:
    def speak(self):
        return "Some sound"

class Dog(Animal):
    pass   

class Cat(Animal):
    pass   

class Cow(Animal):
    pass  

print(Dog().speak())
print(Cat().speak())
print(Cow().speak())
```

# 4. Polymorphism

```python
class Shape:
    def area(self):
        return 0
class Square(Shape):
    def area(self):
        return "Area = side × side"
class Rectangle(Shape):
    def area(self):
        return "Area = length × width"
class Circle(Shape):
    def area(self):
        return "Area = π × r²"

shapes = [Square(), Rectangle(), Circle()]
for s in shapes:
    print(s.area())
```

# 5. Extraction

```python

class Data:
    def extract(self, data):
        return f"Extracting data: {data}"
class Name:
    def extract(self, student_name):
        return f"Extracted name: {student_name}"
class Price:
    def extract(self, price):
        return f"Extracted price: ₹{price}"

print(Name().extract("Akil"))
```
